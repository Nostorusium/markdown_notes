matplotlib在figure(图形)上显示你的数据，这个figure可以视为windows下的控件，Widget。
每个figure包含一个或多个axes，轴(axis的复数)，用来在不同坐标系下展示图形。而axis特指坐标轴。
最简单的创建figure与axes的方法是使用subplots。

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig,ax = plt.subplots()
ax.plot([1,2,3,4],[1,4,2,3])
plt.show()
```

一个figure由许多部分组成并可以设置，它跟踪所有子axes，一组特殊的artist(指如标题，标签等)以及子figure。
你可以用以下方法创建一个figure:

```python
fig0 = plt.figure()          # a empty figure with no axes
fig1,ax1 = plt.subplots()    # a figure with single axes
fig2,ax2 = plt.subplots(2,2) # a figure with 2x2 grid of axes
```

axes指是用于展示数据的区域，存在于figure。它通常包含2个axis(坐标轴)。每个axes有一个标题，有x,y-label。
在figure上，所有可见的东西都是一个artist。它可以是文本信息，2d线条，等等。

有两种思路来使用matplotlib。
1. 显式地创造figure与axes，以面向对象方式将其视为对象来使用他们的方法。
2. 隐式地使用他们，所有操作都依赖pyplot。

```python
# Object-Oriented style
x = np.linspace(0,2,100)
fig,ax = plt.subplots()
ax.plot(x,x,label="y=x")
ax.plot(x,x**2,label="y=x^2")
ax.plot(x,x**3,label="y=x^3")
ax.set_xlabel("X label")
ax.set_ylabel("Y label")
ax.set_title("OOP-style created plot")
ax.legend()
```

```python
# plt-style
plt.figure()
plt.plot(x,x,label="y=x")
plt.plot(x,x**2,label="y=x^2")
plt.plot(x,x**3,label="y=x^3")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("plt-style created plot")
plt.legend()
```
